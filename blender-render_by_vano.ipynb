```python
#@title #<font color=#D2C3AC>Подготовка

from google.colab import drive
drive.mount("/content/drive")

#cache=False#@param{type:"boolean"}
release="4.3"#@param["4.3","4.2","4.1","4.0","3.6","3.5","3.4","3.3","3.2","3.1","3.0","2.93","2.83"]
version="4.3.0"#@param["4.3.0","4.2.3","4.1.1","4.0.2","3.6.17","3.5.1","3.4.1","3.3.21","3.2.2","3.1.2","3.0.1","2.93.18","2.83.20"]
download_blender=False#@param{type:"boolean"}
%cd "/content/drive/MyDrive/"
!mkdir blender-web
%cd "/content/drive/MyDrive/blender-web/"
if download_blender == True:
  !wget -c "https://download.blender.org/release/Blender{release}/blender-{version}-linux-x64.tar.xz"
  !tar -xkvf "blender-{version}-linux-x64.tar.xz"
!chmod +x blender-{version}-linux-x64/blender

from IPython.display import clear_output
clear_output()
```

```python
#@title #<font color=#D2C3AC>Рендер
file_name="room-test"#@param{type:"string"}
device="OPTIX"#@param["CUDA","OPTIX", "CPU"]
animation=True#@param{type:"boolean"}
frame_start=308#@param{type:"number"}
frame_end=600#@param{type:"number"}

if animation==True:
  !./blender-{version}-linux-x64/blender -b "/content/drive/MyDrive/blender-web/{file_name}.blend" -o "/content/drive/MyDrive/blender-web/render/{file_name}/{file_name}" -s {frame_start} -e {frame_end} -a -- --cycles-device {device}
else:
  !./blender-{version}-linux-x64/blender -b "/content/drive/MyDrive/blender-web/{file_name}.blend" -o "/content/drive/MyDrive/blender-web/render/{file_name}/{file_name}" -f {frame_start} -- --cycles-device {device}
```

```python
#@title #<font color=#D2C3AC>Конвертация
zip_archive=False#@param{type:"boolean"}
video=True#@param{type:"boolean"}
FPS=30#@param{type:"number"}
if zip_archive == True:
  !zip {file_name}.zip -r render/{file_name}
  clear_output()
  print("Archive is Done!!!")
if video == True:
  !rm {file_name}.mp4
  !cat render/{file_name}/* | ffmpeg -framerate 30 -i - {file_name}.mp4
  clear_output()
  print("Video is Done!!!")
```
